###  Break , Continue


```python
for i in range(50):
     print(i)
     if i == 5:
        break
```

```python
원금 = 10000
이율 = 0.05
년도 = 0
while i in range(10000000):
    원금 = 원금 + 원금*이율
    년도 += 1
    if 원금 > 20000:
        break
    print('년도 = {}'.format(년도))
    print('현재 원금 = {}'.format(원금))
    print()
    time.sleep(0.5)
```

```python
원금 = 10000
이율 = 0.05
년도 = 0
while i in range(10000000):
    원금 = 원금 + 원금*이율
    년도 += 1

    print('년도 = {}'.format(년도))
    print('현재 원금 = {}'.format(원금))
    print()
    if 원금 > 20000:
        break
    time.sleep(0.5)
```

### continue

```python
for i in range(5):
    if i == 3:
        continue
    print(i)
```

```python
for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
temp = [18,-5,20,50,-17,9]
while temp:
    t=temp.pop(0)
    if t < 0 :
        continue
    print(t)
```
